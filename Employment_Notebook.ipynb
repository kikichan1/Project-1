```python
# Dependencies and Setup
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns

# File to Load
opioid_deaths_to_load = Path('Data/opioid_death_counts.csv')
employment_data_to_load = Path('Data/combined_employment_data.csv')

# # Read Drug Overdose Data File and store into Pandas DataFrames
opioid_csv = pd.read_csv(opioid_deaths_to_load, index_col = 0)
employment_csv = pd.read_csv(employment_data_to_load, index_col = 0)

# merge csv files
employment_vs_opioid_df = pd.merge(employment_csv, opioid_csv, on=['State', 'Year'], how='left')

# pivot dataframe and aggregate
pivot_df = employment_vs_opioid_df.pivot_table(
    index = 'State',
    values = ['Total Unemployed in Civilian Labor Force', 'Total Opioid Deaths'],
    aggfunc = 'sum'
).reset_index()

# calculate % contributions for each state
pivot_df['Unemployment Contribution'] = (pivot_df['Total Unemployed in Civilian Labor Force'] / pivot_df['Total Unemployed in Civilian Labor Force'].sum())  * 100
pivot_df['Opioid Death Contribution'] = (pivot_df['Total Opioid Deaths'] / pivot_df['Total Opioid Deaths'].sum()) * 100

# calculate ratio
pivot_df['Contribution Ratio'] = pivot_df['Opioid Death Contribution'] / pivot_df['Unemployment Contribution']

# contribution table for 
analysis_df = pivot_df[[
    'State',
    'Contribution Ratio',
    'Unemployment Contribution',
    'Opioid Death Contribution'
]]

analysis_df
```

```python

top_10_df = pivot_df.sort_values('Contribution Ratio', ascending=False).head(10)

plt.figure(figsize=(14, 14))
plt.barh(pivot_df['State'], pivot_df['Unemployment Contribution'], color='blue', alpha=0.5, label='Unemployment Contribution')
plt.barh(pivot_df['State'], pivot_df['Opioid Death Contribution'], color='red', alpha=0.5, label='Opioid Death Contribution')
plt.xlabel('Contribution (%)')
plt.title('Comparison of Unemployment and Opioid Death Contributions by State')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Contribution Ratio', y='Opioid Death Contribution', hue='State', data=pivot_df, palette='viridis', s=100)
plt.axvline(1, color='red', linestyle='--', label='Ratio = 1')
plt.title('Contribution Ratio vs. Opioid Death Contribution')
plt.xlabel('Contribution Ratio (Opioid/Unemployment)')
plt.ylabel('Opioid Death Contribution (%)')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(pivot_df.set_index('State')[['Contribution Ratio']], annot=True, cmap='coolwarm', center=1)
plt.title('Heatmap of Contribution Ratio and Opioid Death Contribution')
plt.show()
```
